```python
import pandas as pd
import os
import datetime
from dotenv import load_dotenv
from sqlalchemy import create_engine
from sqlalchemy.types import Integer, Text, String, DateTime, Float
```

```python
df = pd.read_json('phase4_output.json')
df
```

```python
df.drop(columns=['position'], inplace=True)
```

```python
df
```

```python
load_dotenv()
```

```python
DB_PASS = os.getenv("DB_PASS")
DB_USER = os.getenv("DB_USER")
DB_HOST = os.getenv("DB_HOST")
```

```python
engine = create_engine(f"mysql://{DB_USER}:{DB_PASS}@{DB_HOST}/wsb")
conn = engine.connect()
```

```python
df['expiry'] = df['expiry'].apply(datetime.datetime.fromtimestamp)
```

```python
df['expiry']
```

```python
df['close_price'] = df['close_price'].apply(lambda x: -1 if type(x) == str else x)
```

```python
df[df['close_price'] == -1]
```

```python
df['post_date'] = df['post_date'].apply(datetime.datetime.fromtimestamp)
```

```python
df
```

```python
df.to_sql(
    "positions",
    conn,
    if_exists='replace',
    index=False,
    dtype={
        'ticka':String(10),
        'author':String(20),
        'strike':Float,
        'contract':Text,
        'expiry':DateTime,
        'post_date':DateTime,
        'open_price':Float,
        'close_price':Float
    }
)
```
